# **Sistemas Productivos Locales (SPL)**

En esta parte del curso, luego de abordar de manera amplia diferentes conceptos y enfoques, en este espacio, vamos a estudiar algunas técnicas cuantitativas para la identificación inicial de SPL. En este sentido, podemos utilizar una definición inicial de SPL:

"Aglomerados de agentes económicos, políticos y sociales; localizados en un mismo territorio, que presentan vínculos consistentes de articu- lación, interacción, cooperación y aprendizaje; volcados a la introducción de nuevos productos y procesos." (*Cassiolato y Szapiro, 2002, citado por Ayala et al, 2019, p. 42*).

En términos generales, podemos asociar los SPL con:



*   Pequeñas empresas
*   Producción homogénea
*   Interrelaciones entre empresas
*   Aglomeración (vecindad)
*   Demanda de política pública

Las técnicas cuantitativas propuestas en el curso, dan razón a la identificación en el territorio de este tipo de definiciones. Vamos a partir de elementos básicos de estadística descriptiva y luego en técnicas más avanzadas.

En general, utilizaremos datos reales disponibles en el DANE con el que se presente simular un caso practico de identificación inicial de SPL.



# **Fases de la planificación estratégica**

En este ejercicio, seguimos la propuesta metodológica desarrollado por Silva (2003) para la elaboración de estrategias de desarrollo local, este documento pertenece al  **Instituto Latinoamericano y del Caribe de Planificación Económica y Social (ILPES)** de la CEPAL. De acuerdo al autor, en general se pueden establecer cinco (5) fases:

## **Diagnóstico**
Debe proporcionar información que permita conocer la capacidad de desarrollo, las oportunidades y potencialidades, así como los recursos disponibles para ello. En el caso de este ejercicio, se propone arrancer desde los diagnósticos preexistentes de que dispone el respectivo gobierno local, los cuales están basados en información de base secundaria.

## **Vocaciones**
Definidas como la aptitud, capacidad o características especial que tiene la localidad para su desarrollo. Se trata de buscar que es lo que hace especial, propio del lugar, como imagen de marca diferenciada a la localidad para potenciar algunas actividades estratégicas que le permitan impulsar un proceso de desarrollo específico:

*   ¿La ciudad o territorio, tiene vocación productiva, a apartir de la explotación de algunos recursos propios?
*   ¿Tiene vocación turística?
*   ¿Reúne condiciones para ser centro de servicios?

Se recomiendo realizar un análisis desde una perspectiva de áreas de trabajo como: aspectos económico-productivo, socioculturales, infraestructura y servicios públicos, e institucionales. Además se recomienda construir una matriz en la cual se anoten las potencialidades y limitaciones que se encuentran para cada uno de estos sectores, y del análisis de los mismos procurar definir cuales podrían ser vocaciones de la localidad. Se recomienda también adicionar una columna con los principales problemas que se detectan.

## **Objetivos estratégicos y específicos**
Utilizando la técnica de análisis de árboles de problemas, identificar los árboles de medios y fines y la situación esperada que significaría la resolución de los problemas detectados. Los fines y la situación esperada asociados a esta fase se transformaría en los objetivos estratégicos y específicos que se deberían alcanzar.

## **Estrategía local de desarrollo**
La estrategia se define como el camino seleccionado para alcanzar los objetivos propuestos. Se recomienda una técnica de análisis FODA para estudiar las variables internas (fortalezas y debilidades) y externas (oportunidades y amenazas) que pueden condicionar o viabilizar el alcance de los objetivos. **Se debe procurar la definición de la estrategia más adecuada aprovechando al máximo las potencialidades (fortalezas y oportunidades), minimizando los riesgos (fortalezas y amenazas), enfrentando los desafíos (debilidades y amenazas), y procurando neutralizar las limitaciones (debilidades y amenazas)**

## **Proyectos de inversión**
Los proyectos se pueden identificar a partir de los medios del árbol de medios y fines utilizado apra definir los objetivos.

### **Análisis de datos**

**¿Cómo ha evolucionado la producción en Norte de Santander?**

```r
install.packages("pacman")
library(pacman)
```

```r
pacman::p_load(dplyr, ggplot2, tidyr, readxl, plotly, tidyverse)
```

```r
pib <- read_excel("PIB_dpto_.xlsx")
print(pib)

# PIB = Producto Interno Bruto (miles de millones de pesos, año de referencia 2015)
```

```r
ggplot(pib, aes(x = AÑO, y = PIB)) + geom_line() +
  geom_smooth(se=FALSE) +
  labs(
    title = "",
    x = " ",
    y = "PIB (precios constante)"
  )  + theme_bw()
```



*   Tendencia creciente en el largo plazo.
*   Periodos ciclicos con oscilaciones sobre la tendencia de largo plazo, entre 2005 - 2010 por encia de la tendencia, luego el resto de los periodoso muy cercano a la tendencia.



```r
# Tasa de crecimiento del PIB de Norte de Santander

pib$tasa <- (pib$PIB - lag(pib$PIB))/lag(pib$PIB)*100
head(pib)
```

```r
ggplot(pib, aes(x = AÑO, y = tasa)) + geom_line() +
  geom_smooth(se=FALSE) +
    labs(
    title = "",
    x = " ",
    y = "(%)"
  ) +
    theme_bw()
```



*   Los periodos iniciales de la tasa de crecimiento es decreciente, luego converge hacia una especie de banda ente 0 y 5 % a partir de 2010.
*   Luego de 2020 se evidencia un crecimiento prolongado del producto.



**¿Cuál (es) son las actividades económicas con mayor dinámismo en Norte de Santander?**

```r
summary(pib)
```



*   La mayor producción en NS ha sido cercano al 15.168
*   La actividad económica con mayor producción ha sido Comercio, Administración pública y defensa, construcción.
*  No necesariamente los valores máximos expresan la actividad económica con mayor dinámismo pero nos pueden dar una idea inicial de dichos segmentos.



```r
# Seleccionar columnas de interés (puedes agregar más sectores si deseas)

sectores <- pib %>%
  select(AÑO, Agro, Industria, Comercio, Construcción, Sector_Público, Minas)

# Convertir a formato largo

datos_largos <- sectores %>%
  pivot_longer(cols = -AÑO, names_to = "Sector", values_to = "Valor")
```

```r
ggplot(datos_largos, aes(x = AÑO, y = Valor, color = Sector)) +
  geom_line(size = 1.2) +
  geom_point(size = 2) +
  labs(
    title = "Comparación de sectores económicos",
    x = "Año",
    y = "PIB (precios constantes)",
    color = "Sector"
  ) +
  theme_bw()
```

```r
# Calcular tasas de crecimiento por sector
datos_tasas <- pib %>%
  arrange(AÑO) %>%
  mutate(
    tasa_agro = (Agro / lag(Agro) - 1) * 100,
    tasa_industria = (Industria / lag(Industria) - 1) * 100,
    tasa_comercio = (Comercio / lag(Comercio) - 1) * 100,
    tasa_construccion = (Construcción / lag(Construcción) - 1) * 100,
    tasa_sector_publico = (Sector_Público / lag(Sector_Público) - 1) * 100,
    tasa_mina = (Minas / lag(Minas) - 1)* 100
  ) %>%
  select(AÑO, starts_with("tasa_"))

# Convertir a formato largo
datos_largos <- datos_tasas %>%
  pivot_longer(cols = -AÑO, names_to = "Sector", values_to = "Tasa")

# Renombrar sectores para mayor claridad (opcional)
datos_largos$Sector <- recode(datos_largos$Sector,
                              "tasa_agro" = "Agro",
                              "tasa_industria" = "Industria",
                              "tasa_comercio" = "Comercio",
                              "tasa_construccion" = "Construcción",
                              "tasa_sector_publico" = "Sector Público",
                              "tasa_minas" = "Minas")

# Graficar las tasas de crecimiento
ggplot(datos_largos, aes(x = AÑO, y = Tasa, color = Sector)) +
  geom_line(size = 1.2) +
  geom_point(size = 2) +
  labs(
    title = "Tasa de crecimiento por sectores económicos",
    x = "Año",
    y = "Tasa de crecimiento (%)",
    color = "Sector"
  ) +
  theme_bw()
```



*   El sector de minas es mucho mas volatil, esto puede estar relacionado con la exposición al sector externo y los precios internacionales.
*   Los otros sectores parecen tener un comportamiento mucho mas estable

**¿Cómo se relaciona la producción y el valor agregado de las industrias en Norte de Santander?**

```r

mean_prod <- mean(eam$producción)
mean_va <- mean(eam$valor_agregado)

ggplot(eam, aes(x = producción, y = valor_agregado, color = ciiu4)) +
  geom_point() +
  geom_hline(yintercept = mean_va, linetype = "dashed", color = "black") +
  geom_vline(xintercept = mean_prod, linetype = "dashed", color = "black") +
  theme_bw()
```

```r
mean_oci <- mean(eam$ocupados)

ggplot(eam, aes(x = producción, y = ocupados, color = ciiu4)) +
  geom_point() +
  geom_hline(yintercept = mean_oci, linetype = "dashed", color = "black") +
  geom_vline(xintercept = mean_prod, linetype = "dashed", color = "black") +
  theme_bw()
```
